```python
df = pd.read_csv("../input/student-performance-data/student_data.csv")
x = pd.get_dummies(df.drop(['freetime','school'],axis=1))
y = df['school'].apply(lambda x: 1 if x=='GP' else 0)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.2)
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=32, activation='sigmoid', input_dim=len(x_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(x_train,y_train, epochs=1000, batch_size=36)
```

```python
y_hat = model.predict(x_test)
y_hat = ['MS' if val < 0.5 else 'GP' for val in y_hat]
```

```python
accuracy_score(y_test,y_hat)
```
